```python
import tkinter as tk

def button_click(number):
    """Handles button clicks and appends numbers or operations to the input."""
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(0, current + str(number))

def clear():
    """Clears the input field."""
    entry.delete(0, tk.END)

def calculate():
    """Evaluates the expression in the input field."""
    try:
        result = eval(entry.get())
        entry.delete(0, tk.END)
        entry.insert(0, str(result))
    except Exception:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

# Create the main window
root = tk.Tk()
root.title("Calculator")
root.configure(bg="#f8f9fa")  # Light background for the entire calculator

# Entry widget for the calculator display
entry = tk.Entry(root, width=20, borderwidth=5, font=("Arial", 18), justify="right", bg="#e9f5fb", fg="#333")
entry.grid(row=0, column=0, columnspan=4, padx=10, pady=10, sticky="nsew")

# Function to create colorful buttons
def create_button(text, row, column, command=None, bg_color="#ffffff", fg_color="#000000"):
    button = tk.Button(
        root,
        text=text,
        padx=20,
        pady=20,
        font=("Arial", 14),
        bg=bg_color,
        fg=fg_color,
        borderwidth=0,
        command=command,
    )
    button.grid(row=row, column=column, sticky="nsew", padx=2, pady=2)

# Define button colors
number_color = "#dff6f0"  # Light green for numbers
operator_color = "#b3e6ff"  # Light peach for operators
clear_color = "#b3e6ff"  # Soft red for clear
equals_color = "#b3e6ff"  # Light blue for equals

# Create buttons using grid
buttons = [
    ('7', 1, 0, number_color), ('8', 1, 1, number_color), ('9', 1, 2, number_color), ('/', 1, 3, operator_color),
    ('4', 2, 0, number_color), ('5', 2, 1, number_color), ('6', 2, 2, number_color), ('*', 2, 3, operator_color),
    ('1', 3, 0, number_color), ('2', 3, 1, number_color), ('3', 3, 2, number_color), ('-', 3, 3, operator_color),
    ('C', 4, 0, clear_color), ('0', 4, 1, number_color), ('=', 4, 2, equals_color), ('+', 4, 3, operator_color),
]

for text, row, col, color in buttons:
    if text == "=":
        create_button(text, row, col, command=calculate, bg_color=color, fg_color="#ffffff")
    elif text == "C":
        create_button(text, row, col, command=clear, bg_color=color, fg_color="#ffffff")
    else:
        create_button(text, row, col, command=lambda t=text: button_click(t), bg_color=color, fg_color="#333")

# Configure rows and columns for dynamic resizing
for i in range(5):
    root.grid_rowconfigure(i, weight=1)
for i in range(4):
    root.grid_columnconfigure(i, weight=1)

# Start the application
root.mainloop()
```
